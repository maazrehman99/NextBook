```python
import pandas as pd
from IPython.display import display

# Load the first 100 rows to inspect the data types
sample = pd.read_csv("Books.csv", nrows=100)
print("Sample Data Types:")
display(sample.dtypes)

# Use the dtypes from the sample to load the full file
dtypes = sample.dtypes.to_dict()

# Set 'Year-Of-Publication' to str to handle mixed types
dtypes['Year-Of-Publication'] = str

# Read the full file with modified dtypes
p = pd.read_csv("Books.csv", dtype=dtypes, low_memory=False)

# Clean the 'Year-Of-Publication' column
# Convert valid entries to integers and replace invalid entries with NaN
p['Year-Of-Publication'] = pd.to_numeric(p['Year-Of-Publication'], errors='coerce')

# Check for null values
null_counts = p.isnull().sum()
print("Null value counts per column:")
display(null_counts)

# Remove rows with null values in 'Year-Of-Publication'
p_cleaned = p.dropna(subset=['Year-Of-Publication'])

# Number of rows before and after dropping null values
initial_row_count = len(p)
final_row_count = len(p_cleaned)

# Print the number of rows removed and display the cleaned DataFrame
print(f"Number of rows removed: {initial_row_count - final_row_count}")
print("Cleaned DataFrame:")
display(p_cleaned.head())

# If you want to view the entire DataFrame in an Excel-like view in JupyterLab, just output the DataFrame in a cell:
p_cleaned
```

```python
# Check if any data is null in the DataFrame
any_null = p_cleaned.isnull().any().any()

if any_null:
    print("Null values found in the DataFrame:")
    # Print rows with null values
    null_rows = p_cleaned[p_cleaned.isnull().any(axis=1)]
    print(null_rows)

    # Remove rows with null values
    p_cleaned = p_cleaned.dropna()

    print("Null values removed. Updated DataFrame:")
    print(p_cleaned)
else:
    print("There are no null values in the DataFrame.")

```

```python
import pandas as pd

# Load the first 100 rows to inspect the data types
sample = pd.read_csv("final_books_cleaned_data.csv", nrows=100)
print("Sample Data Types:")
print(sample.dtypes)

# Use the dtypes from the sample to load the full file
dtypes = sample.dtypes.to_dict()

# Set 'Year-Of-Publication' to str to handle mixed types
dtypes['Year-Of-Publication'] = str

# Read the full file with modified dtypes
p = pd.read_csv("Books.csv", dtype=dtypes, low_memory=False)

# Clean the 'Year-Of-Publication' column
# Convert valid entries to integers and replace invalid entries with NaN
p['Year-Of-Publication'] = pd.to_numeric(p['Year-Of-Publication'], errors='coerce')

# Remove rows with null or 0 values in 'Year-Of-Publication'
p_cleaned = p.dropna(subset=['Year-Of-Publication'])
p_cleaned = p_cleaned[p_cleaned['Year-Of-Publication'] != 0]

# Convert 'Year-Of-Publication' column to integer
p_cleaned['Year-Of-Publication'] = p_cleaned['Year-Of-Publication'].astype(int)

# Save the final cleaned data to a CSV file
p_cleaned.to_csv("final_cleaned_data.csv", index=False)

# Print the first few rows of the final cleaned data
print("First Few Rows of Final Cleaned Data:")
print(p_cleaned.head())
```

```python
import pandas as pd

# Read the ratings.csv file
ratings = pd.read_csv("ratings.csv")

# Remove rows with null values
ratings_cleaned = ratings.dropna()

# Calculate the sum of ratings for each column
column_sums = ratings_cleaned.sum()

# Sort the columns based on the sums
sorted_columns = column_sums.sort_values(ascending=False)

# Reorder the DataFrame columns based on the sorted order
ratings_sorted = ratings_cleaned[sorted_columns.index]

# Save the cleaned and sorted ratings data to a new CSV file
ratings_sorted.to_csv("cleaned_sorted_ratings.csv", index=False)

# Print the first few rows of the cleaned and sorted ratings data
print("First Few Rows of Cleaned and Sorted Ratings Data:")
print(ratings_sorted.head())
```
