```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data from CSV file
df = pd.read_csv('final_model_data.csv')

# Display basic information about the DataFrame
print(df.info())

# Display the first few rows of the DataFrame
print(df.head())

# Visualize distribution of ratings
sns.countplot(data=df, x='Book-Rating')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()

# Visualize distribution of ratings per user
user_rating_counts = df.groupby('User-ID')['Book-Rating'].count().reset_index(name='Rating Count')
sns.histplot(data=user_rating_counts, x='Rating Count', bins=50)
plt.title('Distribution of Ratings per User')
plt.xlabel('Number of Ratings')
plt.ylabel('Count of Users')
plt.show()

# Visualize distribution of ratings per book
book_rating_counts = df.groupby('Book-Title')['Book-Rating'].count().reset_index(name='Rating Count')
book_rating_counts = book_rating_counts.sort_values(by='Rating Count', ascending=False).head(20)
plt.figure(figsize=(10, 6))
sns.barplot(data=book_rating_counts, y='Book-Title', x='Rating Count', palette='viridis')
plt.title('Top 20 Books by Rating Count')
plt.xlabel('Rating Count')
plt.ylabel('Book Title')
plt.show()
```
